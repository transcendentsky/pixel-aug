```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

thress = [1.5, 2.6, 3.4, 4.3, 5]
values = [0.7654, 0.7019, 0.7103, 0.7013, 0.7648]
thress = np.array(thress)
values = np.array(values)
x = np.arange(0, 7, 0.01)

plt.scatter(thress, values.max() / values)
plt.plot(x, np.clip(1.07-(x-3.4)**2/32, 0, 1))
plt.show()
```

```python
x = np.arange(0, 7, 0.01)
plt.plot(x, np.exp(x/4))
plt.show()
```

```python
x = np.arange(0, 7, 0.01)
plt.plot(x, x**0.63)
plt.show()
```

```python

plt.plot(x, (x.max()-x)**0.1)
plt.show()
```

```python

plt.plot(x, ((x.max()-x)**0.25 + x**0.3)/2)
plt.show()
```
